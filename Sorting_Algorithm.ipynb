```python
def Bubble(l):
    for j in range(len(list(l))):
        for index,i in enumerate(l):
            if index+1==len(l):
                break
            else:
                a=l[index]
                b=l[index+1]
                if a>b:
                    l[index]=b
                    l[index+1]=a
    return l
def Insertion(l):
    for i in range(1,len(list(l))):
        a=l[i]
        j=i-1
        while j>=0 and a<l[j]:
            l[j+1]=l[j]
            j += -1
        l[j+1]=a
    return l
def Merger(l):
    if len(l) >1: 
        mid = len(l)//2 #Finding the mid of the array 
        L = l[:mid] # Dividing the array elements  
        R = l[mid:] # into 2 halves 
  
        Merger(L)  
        Merger(R) #recursive function 
        
        i = j = k = 0
          
        # Copy data to temp arrays L[] and R[] 
        while i < len(L) and j < len(R): 
            if L[i] < R[j]: 
                l[k] = L[i] 
                i+=1
            else: 
                l[k] = R[j] 
                j+=1
            k+=1
          
        # Checking if any element was left 
        while i < len(L): 
            l[k] = L[i] 
            i+=1
            k+=1
          
        while j < len(R): 
            l[k] = R[j] 
            j+=1
            k+=1
    return l
```
